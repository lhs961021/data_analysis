```python
df2 = df.rename(columns={'mosquito_Indicator':'모기발견수','date':'날짜','rain(mm)':'강우량','mean_T(℃)':'평균온도','min_T(℃)':'최저온도','max_T(℃)':'최고온도'})
df2

# column명 변경
```

```python
df2['날짜']= df2['날짜'].str[:7]
df2
# 월별 평균을 내기위한 column 변수 정리
```

```python
df3 = df2.groupby(['날짜']).mean()
df3
 # 월별 평균 값   
```

```python
df2['날짜']= df2['날짜'].str[:4]
df4 = df2.groupby(['날짜']).mean()
df4

#연도별 평균
```

```python
zn = sns.lineplot(x='날짜',y='모기발견수',data=df3)
zn.set_xticklabels(zn.get_xticklabels(),rotation = 90)
```

```python
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 하얀색
 
ax = fig.add_subplot() ## 프래임 생성
 
ax.pie(x=df4['모기발견수'],labels=['2016','2017','2018','2019'],autopct=lambda p : '({:.2f}*10)마리'.format(p)) ## 파이 차트 출력
plt.show()
```
